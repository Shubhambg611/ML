# Learning Matplotlib

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3],[0.5,5,7])
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3],[2,5,7])
plt.title("shubhams data")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```

```python
 from matplotlib import pyplot as plt
 from matplotlib import style

 style.use('ggplot')
 x = [1,5,4]
 y = [25,45,89]
 x2 = [3,7,3]
 y2 = [5,8,9] 
 
plt.plot(x,y)
 plt.plot(x2,y2)
 plt.lagend
 plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style 

style.use('dark_background')
x = [1,5,4]
y = [25,45,89]
x2 = [3,7,3]
y2 = [5,8,9] 
 
plt.plot(x,y,label="first",linewidth=8)
plt.plot(x2,y2,label="second")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style 

style.use('dark_background')
x = [1,5,4]
y = [25,45,89]
x2 = [3,7,3]
y2 = [5,8,9] 
 
plt.scatter(x,y,label="first",linewidth=8)
plt.scatter(x2,y2,label="second")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style 

style.use('dark_background')
x = [1,5,4]
y = [25,45,89]
x2 = [3,7,3]
y2 = [5,8,9] 
 
plt.hist(x,label="first",linewidth=8)
plt.hist(x2,y2,label="second")
plt.legend()
plt.show()
```
